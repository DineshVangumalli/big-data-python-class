## Latest Trends on Twitter
#### To get the latest trends on twitter

```python
api = tweepy.API(auth)
tre = api.trends_place(1) # tre is a list with only one element in it
data = tre[0]            

trends = data['trends']    # Get the trends

names = [trend['name'] for trend in trends]    # Get the name from each trend 
trendsName = '\n'.join(names)
print(trendsName)
```
